```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')
df.to_csv("clean_Superstore.csv", index=False)

#load clean data set
df = pd.read_csv("clean_Superstore.csv")

df.head()
```

```python
pivot_cat = df.pivot_table(
    index='Category',
    values=['Sales', 'Profit'],
    aggfunc='sum'
)
pivot_cat
```

```python
pivot_subcat = df.pivot_table(
    index='Sub-Category',
    values=['Sales', 'Profit'],
    aggfunc='sum'
).sort_values('Profit', ascending=False)

pivot_subcat
```

```python
pivot_seg_reg = df.pivot_table(
    index='Segment',
    columns='Region',
    values=['Sales', 'Profit'],
    aggfunc='sum'
)
pivot_seg_reg
```

```python
print("Most Profitable Category:", pivot_cat['Profit'].idxmax())
print("Least Profitable Category:", pivot_cat['Profit'].idxmin())

print("\nTop 5 Sub-Categories by Profit:")
print(pivot_subcat.head())

print("\nWorst 5 Sub-Categories by Profit:")
print(pivot_subcat.tail())
```

```python
pivot_cat.to_csv("Pivot_Category.csv")
pivot_subcat.to_csv("Pivot_SubCategory.csv")
pivot_seg_reg.to_csv("Pivot_Segment_Region.csv")
```
