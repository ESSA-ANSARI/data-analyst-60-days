```python
import pandas as pd

df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

df.head()

# To know rows and column, shape, dtypes, null values
print("shape: ", df.shape)
print("INFO:")
df.info()
df.describe()
print("NULL values: ")
df.isnull().sum()

# count missing per column and percent missing
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100
pd.DataFrame({'missing_count': missing_count, 'missing_percent': missing_percent}).sort_values('missing_count', ascending=False)

# Are there duplicate rows?
dupes = df.duplicated()
print("Total duplicate rows:", dupes.sum())

# If you want to inspect duplicates:
if dupes.sum() > 0:
    display(df[dupes].head())

df = df.drop_duplicates()

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
df['Ship_Days'] = (df['Ship Date'] - df['Order Date']).dt.days

```
